# How is the rate and approx cost distributed among resturant types?

```python
data['rest_type']=data['rest_type'].astype('str')
```

```python
data['rating']=data['rate'].apply(lambda x:str(x).split('/')[0])
```

```python
data['approx_cost(for two people)']=data['approx_cost(for two people)'].replace('nan',np.nan)
```

```python
data['approx_cost(for two people)']=data['approx_cost(for two people)'].str.replace(',','')
```

```python
data['approx_cost(for two people)']=data['approx_cost(for two people)'].astype('float64')
```

```python
data['approx_cost']=data['approx_cost(for two people)'].apply(lambda x: x/2)
```

```python
data
```

```python
data['rating']=data['rating'].astype('float64')
```

```python
data['rating'].unique()
```

```python
df1=data.loc[0: ,['rating','approx_cost']]
```

```python
df1
```

```python
df2=df1.groupby('rating').mean()
```

```python
df2.head()
```

```python
rating=[a for a,df in df1.groupby('rating')]
plt.plot(rating,df2['approx_cost'])
plt.title("Rating Vs Cost")
plt.xlabel("Resturant_rating")
plt.ylabel("Approx_Cost")
plt.show()
```

```python
sns.relplot(x='rating',y='approx_cost',data=df2)
```
